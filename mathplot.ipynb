```python
import matplotlib.pyplot as plt
import numpy as np

# Veri seti
days = ['Pazartesi', 'Salı', 'Çarşamba', 'Perşembe', 'Cuma', 'Cumartesi', 'Pazar']
temperatures = [22, 24, 19, 23, 25, 20, 18]  # Ortalama sıcaklık değerleri

# Çizgi grafiği oluşturma
plt.plot(days, temperatures, color='b')
# Başlık ve etiketler ekleme
plt.title('Haftalık Sıcaklık Değişimi')
plt.xlabel('Günler')
plt.ylabel('Sıcaklık (°C)')



# Izgara çizgileri ekleme
plt.grid(True)

# Grafiği gösterme
plt.show()
```

```python
import matplotlib.pyplot as plt

# Veri seti
days = ['Pazartesi', 'Salı', 'Çarşamba', 'Perşembe', 'Cuma', 'Cumartesi', 'Pazar']
temperatures_city1 = [22, 24, 19, 23, 25, 20, 18]  # Şehir 1 için sıcaklık değerleri
temperatures_city2 = [25, 27, 22, 26, 28, 23, 21]  # Şehir 2 için sıcaklık değerleri

# Şehir 1 için çizgi grafiği oluşturma
plt.plot(days, temperatures_city1, label='Şehir 1', color='b', marker='o', linestyle='-')

# Şehir 2 için çizgi grafiği oluşturma
plt.plot(days, temperatures_city2, label='Şehir 2', color='r', marker='s', linestyle='--')

# Başlık ve etiketler ekleme
plt.title('Farklı Şehirlerin Haftalık Sıcaklık Değişimi')
plt.xlabel('Günler')
plt.ylabel('Sıcaklık (°C)')

# Açıklama (Legend) ekleme
plt.legend()

# Grafiği gösterme
plt.show()
```

```python
import matplotlib.pyplot as plt

# Veri seti
categories = ['Elektronik', 'Giyim', 'Yiyecek', 'Mobilya', 'Kitap']
sales = [3000, 1500, 2000, 3500, 2500]  # Aylık satışlar

# Çubuk grafiği oluşturma
plt.bar(categories, sales, color='skyblue')

# Başlık ve etiketler ekleme
plt.title('Mağaza Aylık Satışları')
plt.xlabel('Kategoriler')
plt.ylabel('Satış Tutarı (TL)')

# Çubukların üzerine satış tutarlarını ekleme
for i, sale in enumerate(sales):
    plt.text(i, sale+100, str(sale), ha='center')
    
# Grafiği gösterme
plt.show()
```

```python
import matplotlib.pyplot as plt

# Veri seti
categories = ['Pazarlama', 'Ar-Ge', 'Personel', 'Ofis Giderleri', 'Diğer']
budget = [15000, 20000, 30000, 5000, 10000]  # Her bir kategori için bütçe

# Pasta grafiği oluşturma
plt.pie(budget, labels=categories, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink'])

# Başlık ekleme
plt.title('Yıllık Bütçe Dağılımı')

# Grafiği gösterme
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Rastgele veri noktaları oluşturma
np.random.seed(0)  # Rastgelelik için tohum değeri belirleme
x = np.random.rand(100) * 100  # 100 rastgele X değeri (0 ile 100 arasında)
y = np.random.rand(100) * 100  # 100 rastgele Y değeri (0 ile 100 arasında)

# Dağılım grafiği oluşturma
plt.scatter(x, y, color='blue', alpha=0.5, edgecolors='w', s=100)

# Başlık ve etiketler ekleme
plt.title('Rastgele Veri Noktalarının Dağılımı')
plt.xlabel('X Değerleri')
plt.ylabel('Y Değerleri')

# Grafiği gösterme
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Kategorik veriler
categories = ['A', 'B', 'C', 'D', 'E']
# Her kategori için değerler
values = np.random.randint(10, 100, size=len(categories))

# Çubuk grafiği oluşturma
plt.bar(categories, values, color='skyblue')

# Başlık ve etiketler ekleme
plt.title('Kategorik Verilerin Çubuk Grafiği')
plt.xlabel('Kategoriler')
plt.ylabel('Değerler')

# Grafiği gösterme
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Departmanlar ve çalışan sayıları
departments = ['HR', 'Accounting', 'IT', 'Marketing', 'Sales']
employee_counts = [45, 30, 25, 35, 50]

# Çubuk grafiği oluşturma
plt.bar(departments, employee_counts, color='lightgreen')

# Başlık ve etiketler ekleme
plt.title('Şirket Departmanlarındaki Çalışan Sayıları')
plt.xlabel('Departmanlar')
plt.ylabel('Çalışan Sayısı')

# Çubukların üzerine çalışan sayıları ekleme
for i, count in enumerate(employee_counts):
    plt.text(i, count + 1, str(count), ha='center')

# Grafiği gösterme
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Bölümler ve öğrenci sayıları
departments = ['Computer Engineering', 'Electrical Engineering', 'Mechanical Engineering', 'Business Administration', 'Psychology']
student_counts = [120, 100, 90, 150, 80]

# Pasta grafiği oluşturma
plt.pie(student_counts, labels=departments, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

# Başlık ekleme
plt.title('Üniversite Bölümlerindeki Öğrenci Sayıları')

# Grafiği gösterme
plt.show()
```
