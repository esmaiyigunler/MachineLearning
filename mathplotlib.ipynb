```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(100)
data
```

```python
plt.hist(data,bins=10,edgecolor='black')
plt.title("Histogram")
plt.xlabel("Değerler")
plt.ylabel("Frekans")
plt.show()
```

```python
plt.boxplot(data)
plt.title("Box Plot")
plt.ylabel("Değerler")
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame({
 'Group': np.random.choice(['A', 'B', 'C'], size=100),
 'Value': np.random.rand(100)
})

gruplama = df.groupby('Group').mean()
gruplama
```

```python
gruplama.plot(kind = 'line',marker = "o")
plt.title("Gruplara Göre ortalama Değerleri")
plt.xlabel("Grup")
plt.ylabel("Ortalama değerleri")
plt.show()
```

```python
categories = np.array(['A', 'B', 'C', 'D'])
values = np.array([5, 10, 7, 12])
df = pd.DataFrame({
    'Kategori':categories,
    'Değer' : values
})

df
```

```python
plt.bar(df['Kategori'],df['Değer'],color="Red",edgecolor = 'black')
plt.title('Çubuk grafiği')
plt.xlabel("Kategori")
plt.ylabel("Değer")
plt.show()
```

```python
plt.plot(categories,values)
plt.show()
```

```python
Ages= [22, 25, 30, 35, 40, 22, 30, 40, 25, 35, 30, 40]
df = pd.DataFrame({
    'Ages':Ages
})
df
```

```python
plt.hist(df['Ages'],bins=5,color = "black",edgecolor="red")
plt.title("Yaş Histogram")
plt.xlabel("Yaş")
plt.ylabel("değerler")
plt.show()
```

```python
plt.pie(Ages,labels = Ages,autopct = '%1.1f%%',startangle=140)
plt.title("pasta grafiği")
plt.show()
```

```python
Q1= [20, 35, 30, 35]
Q2= [25, 32, 34, 28]
Q3= [30, 34, 37, 30]
Months = ['Jan', 'Feb', 'Mar', 'Apr']
df = pd.DataFrame({
    'Months':Months,
     'Q1':Q1,
     'Q2':Q2,
     'Q3':Q3
})
df
```

```python
plt.bar(df['Months'],df['Q1'],color = 'lightblue',label = Q1)
plt.bar(df['Months'],df['Q2'],bottom = df['Q1'],color = 'salmon',label = Q2)
plt.bar(df['Months'],df['Q3'],bottom = df['Q1'] + df['Q2'],color = 'lightgreen',label = Q3)
```
